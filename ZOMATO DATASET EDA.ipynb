### ZOMATO DATASET EDA

```python
# Importing the necessary libraries
import pandas as pd
import numpy as np
import seaborn as sns  
import matplotlib.pyplot as plt  
```

### DATA PREPARATION

```python
# Importing the datasets required
df=pd.read_csv('Zomato.csv',encoding='ISO-8859-1') # Encoding parameter is specified to read the file
df.head()
```

```python
df.columns # Checking the columns
```

```python
df.describe() # Gives the relation between columns
```

```python
df.isnull().sum() # To check how many null values are present. In this case, there are 9 null values in the column-Cuisines.
```

```python
df.duplicated().sum() # To check the number of duplicates
```

```python
# Loading Country Code Dataset
df2=pd.read_csv('Country-Code.csv')
df2
```

```python
# Now we need to check for common columns in both datasets to merge them together
df.columns
```

```python
df2.columns
```

```python
# Finding the common columns
common_columns = df.columns.intersection(df2.columns)
if not common_columns.empty:
    print("Common columns:", common_columns)
else:
    print("No common columns.")
```

```python
# Merging the two data sets and forming the final data set to be used
df_final=pd.merge(df,df2,on="Country Code")
df_final.head()
```

### DATA VISUALISATION

```python
# Data Visualisation involving pie chart
top_countries = df_final['Country'].value_counts().nlargest(5)
top_countries.plot(kind='pie', autopct='%1.1f%%')
plt.ylabel('')
plt.title('Pie chart for Top 5 Countries')
plt.show()
```

```python
# Using groupby to see the ratings in ascending order
rating = df_final.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index(name='Rating count')
rating.head()
```

```python
# Last values of rating 
rating.tail()
```

```python
# Seeing the different categories within rating
rating['Rating text'].unique()
```

```python
rating
```

```python
# From the rating dataset, we can see that orange is the most given rating, followed by red
colors = ['lightcyan', 'red', 'orange', 'yellow', 'green', 'darkgreen']
sns.countplot(x="Rating color",data=rating,palette=colors)
```

```python
# Finding the top countries with positive rating and its count
Excellent=df_final[df_final["Rating color"]=="Dark Green"].groupby("Country").size().sort_values(ascending=False).reset_index().rename(columns={0:"Rating_count"})
Excellent
```

```python
plt.figure(figsize=(20, 5))
sns.barplot(x="Country", y="Rating_count" ,data=Excellent,palette=colors)
```

```python
sns.lineplot(x='Country',y='Rating_count',data=Excellent,errorbar=None,color='red')
plt.xlabel('Country')
plt.ylabel('Rating_count')
plt.title('Country VS Rating_count')
plt.show()
```

```python
# Finding the top countries with least ratings and their count
poor=df_final[df_final["Rating color"]=="Red"].groupby("Country").size().sort_values(ascending=False).reset_index().rename(columns={0:"Rating_count"})
poor
```

```python
plt.figure(figsize=(15, 5))
sns.barplot(x="Country", y="Rating_count" ,data=poor,palette=colors)
```

```python
sns.lineplot(x='Country',y='Rating_count',data=poor,errorbar=None,color='blue')
plt.xlabel('Country')
plt.ylabel('Rating_count')
plt.title('Country VS Rating_count')
plt.show()
```

```python
# Best city with top rated restaurents
Top=df_final.groupby(["Votes","Restaurant Name","City"]).size().reset_index().tail(50).rename(columns={0:"Rating_count"})
Top
```

```python
plt.figure(figsize=(15, 5))
top_final=Top.groupby(["City"]).size().sort_values(ascending=False).reset_index(name="Rating_count")
sns.barplot(x="City",y="Rating_count", data= top_final,palette=colors)
```

```python
sns.lineplot(x='City',y='Rating_count',data=top_final,errorbar=None,color='purple')
plt.xlabel('City')
plt.ylabel('Rating_count')
plt.title('City VS Rating_count')
plt.show()
```
